```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r"C:\Users\kumar\Downloads\flipkart bigbillion days(laptop)\seperated csv.csv", encoding='latin1')

def category_processor(Processor):
    processor_keywords = ['i3', 'i5', 'i7', 'i9', 'Pentium', 'Athlon', 'APU', 'Ryzen3', 'Ryzen5', 'Ryzen7', 'Ryzen9']
    
    for keyword in processor_keywords:
        if keyword.lower() in Processor.lower():
            return keyword
    return 'Others'

df["ProcessorCategory"] = df['Processor'].apply(category_processor)

plt.figure(figsize=(10, 6))
sns.countplot(x='ProcessorCategory', data=df, order=df['ProcessorCategory'].value_counts().index)
plt.xlabel('Processor Category')
plt.ylabel('Number of Laptops')
plt.title('Distribution of Laptops by Processor Category')
plt.xticks(rotation=45, ha='right')
plt.show()
```
