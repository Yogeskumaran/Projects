```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv("C:\\Users\\kumar\\OneDrive\\Documents\\datasets\\vgsales.csv")
df.head()
df.info()
df.describe()
rowswithnull=df[df.isnull().any(axis=1)]
rowswithnull.info()
selected_region='NA_Sales'
publisher_sales=df.groupby('Publisher')[selected_region].sum().reset_index()
publisher_sales_sorted=publisher_sales.sort_values(by=selected_region,ascending=False)
publishers = 5  
publisher_sales_sorted.head(publishers).plot(
    y=selected_region, kind='pie', figsize=(8, 8), legend=False, labels=publisher_sales_sorted['Publisher'].head(publishers)
)
plt.title(f'Top {publishers} Publishers in {selected_region} Sales')
plt.show()
```
