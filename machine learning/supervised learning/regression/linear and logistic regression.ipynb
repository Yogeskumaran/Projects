```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.datasets import make_classification

# Create synthetic data for linear regression
np.random.seed(0)
X_linear = 2 * np.random.rand(100, 1)
y_linear = 4 + 3 * X_linear + np.random.randn(100, 1)

# Create synthetic data for logistic regression
X_logistic, y_logistic = make_classification(
    n_samples=100, n_features=1, n_informative=1,
    n_redundant=0, n_clusters_per_class=1, random_state=42
)

# Linear Regression
linear_reg = LinearRegression()
linear_reg.fit(X_linear, y_linear)

# Logistic Regression
logistic_reg = LogisticRegression()
logistic_reg.fit(X_logistic, y_logistic)

# Plotting Linear Regression
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.scatter(X_linear, y_linear, color='blue', label='Data')
plt.plot(X_linear, linear_reg.predict(X_linear), color='red', linewidth=3, label='Linear Regression')
plt.title('Linear Regression')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()

# Plotting Logistic Regression
plt.subplot(1, 2, 2)
plt.scatter(X_logistic, y_logistic, color='blue', label='Data')
plt.plot(X_logistic, logistic_reg.predict(X_logistic), color='red', marker='o', linestyle='-', linewidth=2, markersize=8, label='Logistic Regression')
plt.title('Logistic Regression')
plt.xlabel('X')
plt.ylabel('Probability')
plt.legend()

plt.tight_layout()
plt.show()
```
